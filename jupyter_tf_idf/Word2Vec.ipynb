```python
import gensim 
import logging
import math
import os
import pymorphy2
```

```python
files = os.listdir("dictionary")
print("Список документов для анализа")
print("=======================================")
for i in range(len(files)):
    print(str(i)+"=>"+files[i])
print("=======================================")
print("\r\nВведите индекс документа для анализа")
```

```python
documents = []
for j in range(len(files)):
    with open('dictionary/'+files[j]) as f:
        arr = f.readlines()
    documents = arr
for line in documents:
     gensim.utils.simple_preprocess (line)
print (documents)
```

```python
model = gensim.models.Word2Vec (documents, size=150, window=10, min_count=2, workers=10)
model.train(documents,total_examples=len(documents),epochs=10)
```
